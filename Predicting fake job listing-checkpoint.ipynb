```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
print(stopwords.words("english"))
```

```python
stop_words = set(stopwords.words("english"))
```

```python
data['text'] = data['text'].apply(lambda x:x.lower())
```

```python
data['text'] = data['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
data['text'][0]
```

```python
Model_Score = []
Model_Name = []
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test =train_test_split(data.text, data.fraudulent, test_size=0.3)
```

```python
print(X_train.shape) 
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
vect = CountVectorizer()
vect.fit(X_train)
```

```python
X_train_dtm = vect.transform(X_train)
X_train_dtm
```

```python
X_test_dtm = vect.transform(X_test)
X_test_dtm
```

```python
#NaiveBayes Classifier

nb = MultinomialNB()
nb.fit(X_train_dtm, Y_train)
```

```python
Y_pred_nb = nb.predict(X_test_dtm)
```

```python
accuracy_score(Y_test, Y_pred_nb)
```

```python
print ("ROC-AUC:",roc_auc_score(Y_test, Y_pred_nb))
print("Precision:", precision_score(Y_test, Y_pred_nb))
print (classification_report(Y_test, Y_pred_nb))
```

```python
print("Confusion Matrix\n")
print(confusion_matrix(Y_test, Y_pred_nb))
```

```python
plot_confusion_matrix(nb, 
                      X_test_dtm, Y_test, 
                      values_format=' ')
```

```python
Model_Name.append('NaiveBayes')
Model_Score.append(round(roc_auc_score(Y_test, Y_pred_nb), 4))
```

```python
#Decision tree
dt = DecisionTreeClassifier()
dt.fit(X_train_dtm, Y_train)
```

```python
Y_pred_dt = dt.predict(X_test_dtm)
```

```python
accuracy_score(Y_test, Y_pred_dt)
```

```python
print("Precision:", precision_score(Y_test, Y_pred_dt))
print ("ROC-AUC:",roc_auc_score(Y_test, Y_pred_dt))
print (classification_report(Y_test, Y_pred_dt))
```

```python
print("Confusion Matrix\n")
print(confusion_matrix(Y_test, Y_pred_dt))
```

```python
plot_confusion_matrix(dt, 
                      X_test_dtm, Y_test, 
                      values_format=' ')
```

```python
Model_Name.append('Decision Tree')
Model_Score.append(round(roc_auc_score(Y_test, Y_pred_dt), 4))
```
